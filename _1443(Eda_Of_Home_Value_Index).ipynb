```python
import pandas as pd
import matplotlib.pylab as plt
import plotly.express as px
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')  # Changed to 'seaborn-v0_8-whitegrid'

# Use the full option name 'display.max_columns' to avoid ambiguity
pd.set_option("display.max_columns", 500)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()
["color"]

plt.style.use('fivethirtyeight')

from itertools import cycle
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
zstate = pd.read_csv('/content/ZHVI.csv')
zstate = pd.read_parquet('/content/ZHVI.parquet')
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(zstate.corr().sort_values('California'),
            cmap='Blues')
```

```python
zstate[['Montana', 'Connecticut']].plot(figsize=(15, 5),
                                           title='Least Correlated States for House Prices')
plt.show()
```

```python
zstate[['Tennessee', 'Indiana']].plot(figsize=(15, 5),
                                      title='Most Correlated States for House Prices')
plt.show()
```

```python
fig, axs = plt.subplots(17, 3, figsize=(15, 30), sharex=True, sharey=False)
axs = axs.flatten()

for i, state in enumerate(zstate.columns):
  zstate[state].plot(title=state,
                     color=next(color_cycle),
                     ax=axs[i]
                     )
  plt.tight_layout()
  plt.show()
```
